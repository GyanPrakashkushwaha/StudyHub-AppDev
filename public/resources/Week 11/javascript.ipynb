Browser JS vs Node.js
| Method | Description |
| ------------------------ | -------------------------------------------------------------- |
| getElementById() | Selects an element by its ID |
| getElementsByClassName() | Selects elements with a given class name (live HTMLCollection) |
| getElementsByTagName() | Selects elements with a given tag name (live HTMLCollection) |
| getElementsByName() | Selects elements with a given name attribute |
| querySelector() | Returns the first match for a CSS selector |
| querySelectorAll() | Returns all matches for a CSS selector (static NodeList) |
| createElement() | Creates a new element node |
| createTextNode() | Creates a new text node |
| appendChild() | Appends a node as the last child |
| removeChild() | Removes a child node |
| replaceChild() | Replaces a child node |
| insertBefore() | Inserts a node before another node |
| cloneNode() | Clones a node |
| addEventListener() | Attaches an event handler |
| removeEventListener() | Removes an event handler |
| setAttribute() | Sets an attribute value |
| getAttribute() | Gets an attribute value |
| removeAttribute() | Removes an attribute |
| classList.add() | Adds one or more classes to an element |
| classList.remove() | Removes one or more classes from an element |
| classList.toggle() | Toggles a class on an element |
| setAttributeNode() | Adds a new attribute node to an element |
| normalize() | Joins adjacent text nodes and removes empty ones |
| contains() | Checks if a node is a descendant of another |
| focus() | Sets focus to an element |
| blur() | Removes focus from an element |
| innerHTML(property) | Gets or sets the HTML content of an element |
| textContent(property) | Gets or sets the text content of an element |
